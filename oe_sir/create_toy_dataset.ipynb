```python
import pandas as pd
import os
import shutil
from tqdm import tqdm
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/namnv78/similaritymatching/tuha5/facebook_matching/src/df_train.csv")
```

```python
# !tar -xf "/content/drive/MyDrive/Viettel DTalent/ViettelTalentsPhase2/Phan Nhật Minh/Datasets/Copy of query_images.tar.gz"
# !tar -xf "/content/drive/MyDrive/Viettel DTalent/ViettelTalentsPhase2/Phan Nhật Minh/Datasets/reference_images_4.tar.gz"
```

```python
os.mkdir('toy_ds')
os.mkdir('toy_ds/query_images')
os.mkdir('toy_ds/reference_images')
```

```python
def remove_folder(folder):
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        try:
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)
        except Exception as e:
            print('Failed to delete %s. Reason: %s' % (file_path, e))

def process(part=1):
    !tar -xf "/content/drive/MyDrive/Viettel DTalent/ViettelTalentsPhase2/Phan Nhật Minh/Datasets/reference_images_{part}.tar.gz"
    for index, row in tqdm(df.iterrows()):
        query = row['query_id']
        reference = row['reference_id']

        inference_id = int(reference[1:])
        if part*100000 <= inference_id < (part+1)*100000:
            shutil.copyfile('reference_images/R{}.jpg'.format(inference_id), 'toy_ds/reference_images/R{}.jpg'.format(inference_id))
            shutil.copyfile('query_images/{}.jpg'.format(query), 'toy_ds/query_images/{}.jpg'.format(query))
    
    remove_folder("reference_images")
```

```python
!tar -xf "/content/drive/MyDrive/Viettel DTalent/ViettelTalentsPhase2/Phan Nhật Minh/Datasets/Copy of query_images.tar.gz"
```

```python
for idx in [4]:
    process(idx)
```
